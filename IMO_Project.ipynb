```python
import json
#import time

fd = open("IMO_record.json", 'r')

txt = fd.read()

fd.close()

# print(txt)

dct_record = json.loads(txt)

# dct_record


print("*****************************************************************************************")
print("***************************************MENU**********************************************")

for i in dct_record:
    print(i, dct_record[i])
    
print("")
print("")



will = input("What do you want to do? (add_in_inventory/purchase)")
if will.lower() == 'purchase':
    

    ui_prod = input("Enter the product's ID you want to buy: ")
    ui_quat = int(input("Enter the quantity of product you want: "))

    if ui_prod in dct_record.keys():
        
        if int(dct_record[ui_prod]['quan']) >= ui_quat:
            print(f"Name of the product is {dct_record[ui_prod]['name']}")
            print(f"Product {dct_record[ui_prod]['name']} costs {dct_record[ui_prod]['price']} Rs. each ")
            print(f"We have {dct_record[ui_prod]['quan']} quantity available of '{dct_record[ui_prod]['name']}'")
            print("******************************************************************************************")
            print(f"Your total bill is : {int(dct_record[ui_prod]['price']) * ui_quat}")
        
            dct_record[ui_prod]['quan'] = str(int(dct_record[ui_prod]['quan']) - ui_quat)
        
        else:
            print(f"Sorry but we do not have that much stock left for {dct_record[ui_prod]['name']} ")
            print(f"We only have {dct_record[ui_prod]['quan']} stock left")
            choice = input(f"Will you pleasee reduce your qunatity to {dct_record[ui_prod]['quan']} or less? (Yes/No)")
            
            if choice.lower() == 'yes':
                ui_revised_quat = input(f"Enter the new quntity you want (but please keep it below {dct_record[ui_prod]['quan']})")
                print(f"Name of the product is {dct_record[ui_prod]['name']}")
                print(f"Product {dct_record[ui_prod]['name']} costs {dct_record[ui_prod]['price']} Rs. each ")
                print(f"We have {dct_record[ui_prod]['quan']} quantity available of '{dct_record[ui_prod]['name']}'")
                print("******************************************************************************************")
                print(f"Your total bill is : {int(dct_record[ui_prod]['price']) * int(ui_revised_quat)}")
        
                dct_record[ui_prod]['quan'] = str(int(dct_record[ui_prod]['quan']) - int(ui_revised_quat))
                
                print("Item successfully purchased")
        report = input("Would you like to generate sales report? (yes/no)")
        
        if report.lower() == 'yes':
            sales = {}

            sales[str((len(sales) + 1))] = {'Id': ui_prod, 'name': dct_record[ui_prod]['name'], 'quantity': ui_quat, 'price': dct_record[ui_prod]['price']}

            jsa = json.dumps(sales)
            fd = open("sales.json", 'a')
            fd.write(jsa)
            fd.close()
        
        print("Reports are created in sales.json")    
        
        print("Thank you for visiting!!")
    

    else:
        print(f"Product with ID {ui_prod} does not exist :(")


else:
    
    a = input("You want to add products in inventory, right? (yes/no): ")
    
    if a.lower() == 'yes':
    
        prod_id = input("Enter product ID: ")
        name = input("Enter name of product: ")
        price = input("Enter price of product: ")
        quan = input("Enter quantity of product: ")
        dct_record[prod_id] = {'name': name, 'price': price, 'quan': quan}
        if (prod_id in dct_record.keys()) and (name == dct_record[prod_id]['name']) and (price == dct_record[prod_id]['price']):
            dct_record[prod_id]['quan'] =  str(int(dct_record[prod_id]['quan']) + int(quan))
            
       
        print("product successfully added into inventory!")
    
    else:
        
        pass
    
js = json.dumps(dct_record)

fd = open("IMO_record.json", 'w')

fd.write(js)

fd.close()


```
